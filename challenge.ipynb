  ## Identifying Outliers using Standard Deviation

```python
# initial imports
import pandas as pd
import numpy as np
import random
from sqlalchemy import create_engine
```

```python
# create a connection to the database
engine = create_engine("postgresql://postgres:password@localhost:5432/Transaction")
```

```python
# code a function to identify outliers based on standard deviation
query_1="Select holder.id,trn.amount from card_holder holder join credit_card card on holder.id=card.id_card_holder join transactions trn on trn.card=card.card"
transactions=pd.read_sql(query_1, engine)
elements=np.array(transactions.amount)
mean = np.mean(elements, axis=0)
sd = np.std(elements, axis=0)
cut_off = sd * 3
lower=  mean - cut_off
upper = mean + cut_off
```

```python
# find anomalous transactions for 3 random card holders
outliers = [x for x in elements if x < lower or x > upper]
outliers_df = pd.DataFrame(data=outliers)
outliers_df
```

  ## Identifying Outliers Using Interquartile Range

```python
# code a function to identify outliers based on interquartile range
def outlier_treatment(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
lowerbound,upperbound = outlier_treatment(transactions.amount)
```

```python
# find anomalous transactions for 3 random card holders
transactions[(transactions.amount < lowerbound) | (transactions.amount > upperbound)]
```
